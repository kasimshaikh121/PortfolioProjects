```python
#Importing Libraries 

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt 

import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure  

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None 


#Reading Data

df = pd.read_csv(r'C:\Users\Hp\Downloads\movies.csv')

```

```python
# Looking at Data 
df.head()
```

```python
# Looking for missing data 

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,round(pct_missing*100)))
```

```python
# To know Data Types for the Columns in the dataset 

print(df.dtypes)
```

```python
# Changing Data Types for columns



# If error prompts "IntCastingNanError", use statement below
# Occurs if there is missing values in the dataset columns

df['budget'] = pd.to_numeric(df['budget'], errors='coerce').fillna(0).astype(int)
df['gross'] = pd.to_numeric(df['gross'], errors='coerce').fillna(0).astype(int) 


print(df.dtypes)
```

```python
# Re-arranging the order based on Gross Value in descending order 

df = df.sort_values(by=['gross'], inplace=False, ascending=False)

df
```

```python
# Drop any duplicates 

df.drop_duplicates()
```

```python
# Budget high correlation 
```

```python
#Scatter plot of Budget vs Gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show()
```

```python
df.head()
```

```python
# Plotting Budget vs Gross Earnings using Seaborn 

sns.regplot(x='gross', y='budget', data=df, scatter_kws={"color":"blue"}, line_kws={"color":"red"})
```

```python
# Looking at correlation of data containing numerical data-types

numeric_df = df.select_dtypes(include=['number'])

numeric_df.corr()
```

```python
# Correlation Matrix between all numeric columns

numeric_df.corr(method='pearson')  # Pearson, Kendall, Spearman  
```

```python
numeric_df.corr(method='kendall')
```

```python
numeric_df.corr(method='spearman')
```

```python
# High correlation between Budget and Gross 

correlation_matrix = numeric_df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix of Numerical Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Look at Company
df.head()
```

```python
df_numerized = df

for col_names in df_numerized.columns:
    if(df_numerized[col_names].dtypes == 'object'):
        df_numerized[col_names] = df_numerized[col_names].astype('category')
        df_numerized[col_names] = df_numerized[col_names].cat.codes
        
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix of Numerical Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
pd.set_option('display.max_rows', None)

correlation_mat = df_numerized.corr(method='pearson')

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs 
```

```python
# Viewing high correlations 

high_correlation = sorted_pairs[sorted_pairs>0.5]

high_correlation
```
